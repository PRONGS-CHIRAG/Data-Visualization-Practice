```python
#setup dependencies
import pandas as pd
import matplotlib.pyplot as plt 
%matplotlib inline
import seaborn as sns
```

```python
#read data
fifa_fp = "Datasets/fifa.csv"
spotify_fp = "Datasets/spotify.csv"
museum_fp = "Datasets/museum_visitors.csv"
fifa_data = pd.read_csv(fifa_fp,index_col="Date",parse_dates=True)
spotify_data = pd.read_csv(spotify_fp,index_col="Date",parse_dates=True)
museum_data = pd.read_csv(museum_fp,index_col="Date",parse_dates=True)
```

```python
fifa_data.head()
```

```python
#plotting figures
plt.figure(figsize=(20,10))
sns.lineplot(data=fifa_data)
```

```python
spotify_data.head()
```

```python
plt.figure(figsize=(20,10))
plt.title("Daily Global Streams of Popular Songs in 2017-2018")
sns.lineplot(data=spotify_data)
```

```python
#plotting individual lineplots
plt.figure(figsize=(14,6))
plt.title("Daily Global Streams of Popular Songs in 2017-2018")
sns.lineplot(data=spotify_data['Shape of You'],label="Shape of You")

sns.lineplot(data=spotify_data['Despacito'],label="Despacito")

sns.lineplot(data=spotify_data['HUMBLE.'],label="Humble")
plt.xlabel("Date")
```
