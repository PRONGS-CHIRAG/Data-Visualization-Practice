```python
#setup dependencies
import pandas as pd
import matplotlib.pyplot as plt 
%matplotlib inline
import seaborn as sns
```

```python
# filepaths and loading data 
ins_fp="Datasets/insurance.csv"
ins_data = pd.read_csv(ins_fp)
```

```python
ins_data
```

```python
#drawing scatterplot for bmi and charges
plt.title("Scatterplot of BMI vs Charges")
sns.scatterplot(x=ins_data['bmi'],y=ins_data['charges'])
```

```python
#adding regression line to check strength of relationship
sns.regplot(x=ins_data['bmi'],y=ins_data['charges'])
```

```python
#using scatterplot to plo 3 criteria 
sns.scatterplot(x=ins_data['bmi'],y=ins_data['charges'],hue=ins_data['smoker'])
```

```python
#lmplot being used to add 2 regression lines to the recent scatterplot
sns.lmplot(x='bmi',y='charges',hue='smoker',data=ins_data)
```

```python
#plotting categorical scatter plot
sns.swarmplot(x=ins_data['smoker'],y=ins_data['charges'])
```
