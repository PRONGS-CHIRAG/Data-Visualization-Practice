```python
#setup dependencies
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
#dataset paths and loading data
flight_fp="Datasets/flight_delays.csv"
flight_data = pd.read_csv(flight_fp,index_col="Month")
```

```python
flight_data
```

```python
#drawing barchart
plt.figure(figsize=(14,6))
plt.title("Average Delay for Spirit Airlines flights by month")
sns.barplot(x=flight_data.index,y=flight_data['NK'])
plt.xlabel("Months")
plt.ylabel("Arrival delay (in minutes)")
```

```python
#drawing heatmap
plt.figure(figsize=(14,6))
plt.title("Average Delay for Spirit Airlines flights by month")
sns.heatmap(data=flight_data,annot=True)
plt.xlabel("Airline")
plt.ylabel("Months")
```
