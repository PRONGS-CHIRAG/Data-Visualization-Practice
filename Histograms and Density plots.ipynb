```python
#importing libraries
import pandas as pd
import matplotlib.pyplot as plt 
%matplotlib inline
import seaborn as sns
```

```python
iris_fp="Datasets/iris.csv"
iris_data = pd.read_csv(iris_fp,index_col="Id")
iris_data.head()
```

```python
#draw histogram
sns.histplot(iris_data['Petal Length (cm)'])
```

```python
#Density plots aka Kernel density estimate is like a smoothed histogram
sns.kdeplot(data=iris_data['Petal Length (cm)'],shade=True)
```

```python
#kde plots can have 2 dimensions also
sns.jointplot(x=iris_data['Petal Length (cm)'],y=iris_data['Sepal Width (cm)'],kind="kde")
```

```python
#color coding histograms
sns.histplot(data=iris_data,x='Petal Length (cm)',hue='Species')
plt.title("Histogram of Petal Lengths, by species")
```

```python
#color coded density plots
sns.kdeplot(data=iris_data,x='Petal Length (cm)',hue='Species')
plt.title("Distribution of Petal Length by species")
```
