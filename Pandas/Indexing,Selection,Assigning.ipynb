```python
wine_data['country']# another way to access country column
```

```python
''' 
Indexing in Pandas:
There are two ways of indexing in pandas :
Index-based selection(iloc) - selecting data based on its numerical position in data
Label-based selection(loc) - selecting data based on its label and not its position
df.loc[0:1000] returns 1001 entries
df.iloc[0:1000] returns 1000 entries
'''
```

```python
wine_data.iloc[0] # retrieves the first entry
```

```python
wine_data.iloc[:,0] # retrieves first column
wine_data.iloc[:3,0] # retrieves first column entries for first 3 rows
```

```python
wine_data.iloc[[0,1,2,3,4,5,6],0]# lists can also be passed
```

```python
wine_data.iloc[-5:]#prints last 5 rows
```

```python
wine_data.loc[0,'country'] # prints country of first row
```

```python
#conditional selection

#check if the wine is italian or not
wine_data.country == 'Italy'
```

```python
#doing the same with loc
wine_data.loc[wine_data.country== 'Italy']
```

```python
wine_data.loc[(wine_data.country == 'Italy') & (wine_data.price > 50)]
```

```python
wine_data.loc[(wine_data.country == 'Italy') | (wine_data.price > 50)]
```

```python
wine_data.loc[wine_data.country.isin(['Italy','India'])]
```
